```python

bio_data = {
    'Name':[],
    'Gender':[],
    'Matric_no':[],
    'Phone_number':[],
    'Faculty':[],
    'Department':[],
    'Level':[],
    'Semester':[],
    'Batch':[],
    'Group':[],
    'Course_registered':[],
    'Program':[],
    'Course_title':[],
    'Course_code':[],
    'Course_units':[],
    'Score':[],
    'Grade':[],
    'Points':[],
    'Remarks':[],
    'Total_points':[],
    'Total_units':[],
    'GPA':[]
}


print('student bio_data')
print('==========')
Name = input('Enter your name: ')
bio_data['Name'].append(Name)

Gender = input('Enter your gender: ')
if Gender == 'male':
    print('M')
elif Gender == 'female':
    print('F')
else:
    print('invalid,try again')
    Gender = input('Enter your gender: ')
bio_data['Gender'].append(Gender)

Matric_no = input('Enter your matric no: ')
bio_data['Matric_no'].append(Matric_no)

Phone_number = input('Enter your phone_number: ')
while True:
    if Phone_number.isdigit() and len(Phone_number) == 11:
        break
    else:
        print('Error, Try again')
        Phone_number = input('Enter your phone_number: ')
        bio_data['Phone_number'].append(Phone_number)

Faculty = input('Enter your Faculty: ')
bio_data['Faculty'].append(Faculty)

 
Department = input('Enter your Department: ')
bio_data['Department'].append(Department)

Level = int(input('Enter your Level: '))
while True:
    if Level in [100,200,300,400,500]:
        break
    else:
        print('The above level is invalid ')
        Level =int(input('Enter your Level`: '))
        bio_data['Level'].append(Level)

Semester = input('Enter your Semester:')
if Semester =='HARMATTAN':
    print('H') 
elif Semester == 'RAIN':
        print('R')
else:
    print('invalid, try again')
    Semester = input('Enter your Semester: ')
    bio_data['Semester'].append(Semester)
    
Batch = input('Enter your batch: ')
bio_data['Batch'].append(Batch)

Group = input('Enter your group: ')
bio_data['Group'].append(Group)

Program = input('Enter your program: ')
bio_data['Program'].append(Program)

Course_registered  = int(input('Enter the Num_courses : '))
while True:
    if Course_registered in [1,2,3,4,5,6]:
        break
    else:
        print('You cannot register more than six courses')
        Course_registered  = int(input('Enter the Num_courses: '))
        bio_data['course_registered'].append(course_registered)
    
    
Total_points = 0
Total_units = 0

def grading(Score):
    if Score>=70 and Score<=100:
        Grade = 'A'
        bio_data['Grade'].append(Grade)
        Points = 5 * Course_units
        bio_data['Points'].append(Points)
        Remarks = 'PASSED'
        bio_data['Remarks'].append(Remarks)
    elif Score >=60 and Score<=69:
        Grade = 'B'
        bio_data['Grade'].append(Grade)
        Points = 4 * Course_units
        bio_data['Points'].append(Points)  
        Remarks = 'PASSED'
        bio_data['Remarks'].append(Remarks)
    elif Score>=50 and Score <=59:
        Grade = 'C'
        bio_data['Grade'].append(Grade)
        Points = 3 * Course_units
        bio_data['Points'].append(Points)  
        Remarks = 'PASSED'
        bio_data['Remarks'].append(Remarks)
    elif Score >=45 and Score<=49:
        Grade = 'D'
        bio_data['Grade'].append(Grade)
        Points = 2 * Course_units
        bio_data['Points'].append(Points)  
        Remarks = 'PASSED'
        bio_data['Remarks'].append(Remarks)
    elif Score>=40 and Score<=44:
        Grade = 'E'
        bio_data['Grade'].append(Grade)
        Points = 1 * Course_units
        bio_data['Points'].append(Points)  
        Remarks = 'PASSED'
        bio_data['Remarks'].append(Remarks)
    else:
        Grade = 'F'
        bio_data['Grade'].append(Grade)
        Points = 0 * Course_units
        bio_data['Points'].append(Points)  
        Remarks = 'PASSED'
    return Grade, Points, Remarks
for i in range(1, Course_registered+1):
    Course_title = input('Enter your Course_title : ')
    bio_data['Course_title'].append(Course_title)
    
    Course_code = input('Enter your course_code : ')
    bio_data['Course_code'].append(Course_code)
    
    Course_units = int(input('Enter your Course_units : '))
    bio_data['Course_units'].append(Course_units)
    
   
    Score = int(input('Enter your Score : '))
    bio_data['Score'].append(Score)
    
    Grade, Points, Remarks = grading(Score)
    bio_data['Grade'].append(Grade)
    bio_data['Points'].append(Points)
    bio_data['Remarks'].append(Remarks)
    
Total_units += Course_units
        
Total_points += Points
    
bio_data['Total_units'].append(Total_units)
bio_data['Total_points'].append(Total_points)

print('The Total_units is: ', Total_units)
print('The Total_points is: ', Total_points)

GPA = (Total_points/Total_units)
bio_data['GPA'].append(GPA)
print('Your GPA is: ', GPA)
if GPA >= 4.50 and GPA <= 5.00:
    degree_class = 'First class'
elif GPA >= 3.50 and GPA <= 4.49:
    degree_class = 'Second class upper'
elif GPA >= 2.50 and GPA <= 3.49:
    degree_class = 'Second class lower'
elif GPA >= 1.50 and GPA <= 2.49:
    degree_class = 'Third class'
else:
    degree_class = 'pass' 
    
def results():
    print(f"""
                  *************************************************************************
                        LADOKE AKINTOLA UNIVERSITY OF TECHNOLOGY,OGBOMOSO, OYO STATE
                        
                                        FACULTY OF {Faculty}
                                      DEPARTMENT OF {Department}
                                    
                                        STUDENT STATEMENT OF RESULT
                   ************************************************************************
                    
     """)
    
    print(f"""
   ------------------------------------------------------------------------
   | MATRIC NO | {Matric_no} | STUDENT NAME | {Name}                       
   |-----------------------------------------------------------------------
   |GENDER     | {Gender}    | PHONE NUMBER | {Phone_number}               
   |-----------------------------------------------------------------------
   | PROGRAM | {Program} | BATCH | {Batch} | GROUP |{Group}                
   ------------------------------------------------------------------------
    ------------------------------------------------------------------------------------------------------------------
          
    S/N | COURSE TITLE    | COURSE CODE    | COURSE UNIT    | SCORE    | GRADE     | POINT         |  REMARKS
    ------------------------------------------------------------------------------------------------------------------
    """)
    
    for i in range(Course_registered):
        print(f"|{i+1}|{bio_data['Course_title'][i]:<20}|{bio_data['Course_code'][i]:<15}|{bio_data['Course_units'][i]:<5}|{bio_data['Score'][i]:<3}|{bio_data['Grade'][i]:<3}|{bio_data['Points'][i]:<2}|{bio_data['Remarks'][i]:<3}")
        print(f"|                                                                                                                                                                                                                    ")
        print(f"|----------------------------------------------------------------------------------------------------")
    print(f"""                            STUDENT RESULT 
   |- ---------------------------------------------------------------------------------|
   |                                 {Semester} SEMESTER PERFORMANCE                   |    
   |-----------------------------------------------------------------------------------|
   |     POINT               |        UNIT        |    GPA       |  CLASS              |
   |-----------------------------------------------------------------------------------|
   |  {Total_points:<13}     |  {Total_units:<11} | {GPA:<9.2f}  |  {degree_class:<15} |
   | ----------------------------------------------------------------------------------|
   |                                                                                   |
   |-----------------------------------------------------------------------------------|
   | EXAMINATION OFFICER | A.G HEAD OF DEPARTMENT | DEAN OF FACULTY                    |
   |---------------------|------------------------|------------------------------------|
   | DR R.O ADEPOJU      | DR A.A OLANIYI         | PROF.A.T.J OGUNKUNLE               |
   -------------------------------------------------------------------------------------
    """)          
              
        
results()

```
